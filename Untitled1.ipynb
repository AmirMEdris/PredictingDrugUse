```python
import pickle as pkl 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score,confusion_matrix,accuracy_score
```

```python
vtc = pkl.load(open('vtc.pkl','rb'))
```

```python
stimvtc = vtc
haluvtc = vtc
depvtc = vtc
```

```python
df = pd.read_csv('prepeddata.csv',index_col='Unnamed: 0')
df['stimulant'] = df['stimulant'].apply(lambda x: 'None' if x == 0 else 'low' if x <= 2 else 'high')
df
df['hallucinagen'] = df['hallucinagen'].apply(lambda x: 'None' if x == 0 else 'low' if x <= 2 else 'high')
df
df['depressant'] = df['depressant'].apply(lambda x: 'None' if x == 0 else 'low' if x <= 2 else 'high')
df
```

```python
X = df.drop(['stimulant','hallucinagen','depressant'],axis = 1)
stimy = df.stimulant
depy = df.depressant
hally = df.hallucinagen
stimXtrain,stimXtest,stimYtrain,stimYtest = train_test_split(X,stimy)
depXtrain,depXtest,depYtrain,depYtest = train_test_split(X,depy)
hallXtrain,hallXtest,hallYtrain,hallYtest = train_test_split(X,hally)
```

```python
stimvtc.fit(stimXtrain,stimYtrain)
stimtrain=stimvtc.predict(stimXtrain)
stimtest=stimvtc.predict(stimXtest)
depvtc.fit(depXtrain,depYtrain)
deptrain=depvtc.predict(depXtrain)
deptest=depvtc.predict(depXtest)
haluvtc.fit(hallXtrain,hallYtrain)
halutrain=haluvtc.predict(hallXtrain)
halutest=haluvtc.predict(hallXtest)
```

```python
f1_score(stimYtrain,stimtrain,average='weighted'),f1_score(stimYtest,stimtest,average='weighted')
confusion_matrix(stimYtest,stimtest)
```

```python
f1_score(depYtrain,deptrain,average='weighted'),f1_score(depYtest,deptest,average='weighted')
```

```python
f1_score(hallYtrain,halutrain,average='weighted'),f1_score(hallYtest,halutest,average='weighted')
```
