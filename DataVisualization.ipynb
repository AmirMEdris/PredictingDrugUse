```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('prepeddata.csv',index_col='Unnamed: 0')
```

```python
df.groupby('stimulant')[['Alchol','Nicotine','Chocalate','Caff',]].mean().plot.bar()
plt.savefig('LesserStimulantUsageByStimulant.png', bbox_inches='tight')
```

```python
df.groupby('Impulsive')[['Nscore','Escore','Oscore','Ascore','Cscore']].mean().plot()
```

```python
df.groupby('stimulant')[['Nscore','Escore','Oscore','Ascore','Cscore']].mean().plot.bar()
plt.savefig('PersonalityScoresByStimulant.png', bbox_inches='tight')
```

```python
df.groupby('stimulant')[['Alchol','Caff','Chocalate','Nicotine']].mean().plot.bar()
plt.savefig('SecondaryDrugUseAmongHighStimPatients.png', bbox_inches='tight')
```

```python
sns.distplot(df.stimulant,label='Stimulant Use')
sns.distplot(df.hallucinagen,label = 'Hallucinagen Use')
sns.distplot(df.depressant,label = 'Depressant Use')


plt.legend()
```

```python
sns.distplot(df.Cscore,label = 'Careful')
```

```python
sns.distplot(df.Nscore,label='Nscore')
sns.distplot(df.Ascore,label = 'Ascore')
sns.distplot(df.Oscore,label = 'Oscore')

plt.legend()
```

```python

df.groupby('stimulant')[['Impulsive','SensationSeeking']].mean().plot.bar()
plt.savefig('ControlRelatedStatsByStimUse.png', bbox_inches='tight')
```

```python
df.groupby('Nscore')[['depressant','stimulant','hallucinagen']].mean().plot()

plt.savefig('DrugUseByHowAnxious.png', bbox_inches='tight')
```

```python
df.groupby('Ascore')[['depressant','stimulant','hallucinagen']].mean().plot()
plt.savefig('DrugUseByHowCooperativeYouAre.png', bbox_inches='tight')
```

```python
df.groupby('Oscore')[['depressant','stimulant','hallucinagen']].mean().plot()
plt.savefig('DrugUseByHowOpenToNewExperiencesYouAre.png', bbox_inches='tight')
```

```python
df.groupby('Cscore')[['depressant','stimulant','hallucinagen']].mean().plot()
plt.savefig('DrugUseByHowCarefulYouAre.png', bbox_inches='tight')
```

```python
df.groupby('Escore')[['depressant','stimulant','hallucinagen']].mean().plot()
plt.savefig('DrugUseByHowExtravertedYouAre.png', bbox_inches='tight')
```

```python
sns.distplot(df.Impulsive)
sns.distplot(df.SensationSeeking)
plt.savefig('ControlParams.png', bbox_inches='tight')

```

```python
sns.distplot(df.Nscore,label='Nscore')
sns.distplot(df.Ascore,label = 'Ascore')
sns.distplot(df.Oscore,label = 'Oscore')
sns.distplot(df.Cscore,label = 'Cscore')
sns.distplot(df.Escore,label = 'Escore')

plt.legend()
plt.savefig('NormalDistPersonalityScoresParams.png', bbox_inches='tight')
```

```python
sns.distplot(df.stimulant,label='Stimulant Use')
sns.distplot(df.hallucinagen,label = 'Hallucinagen Use')
sns.distplot(df.depressant,label = 'Depressant Use')


plt.legend()
plt.savefig('PoissonDistOfDrugUsers.png', bbox_inches='tight')
```
