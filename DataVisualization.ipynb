```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('prepeddata.csv',index_col='Unnamed: 0')
df['stim'] = df['stimulant'].apply(lambda x: 'low' if x <= 2 else 'high')
df
```

```python
df.groupby('stim')[['Alchol','Nicotine','Chocalate','Caff']].mean().plot.bar()
plt.savefig('pics/LesserStimulantUsageByStimulant.png', bbox_inches='tight')
```

```python
for i in ['Nscore','Escore','Oscore','Ascore','Cscore','Impulsive','SensationSeeking']:
    sns.lineplot(data=df.groupby(i)[['Alchol','Nicotine','Chocalate','Caff']].mean())
    plt.show()
```

```python
for i in ['Alchol','Nicotine','Chocalate','Caff']:
    sns.lineplot(data=df.groupby(i)[["stimulant","depressant","hallucinagen"]].mean())
    plt.show(close=True)
```

```python

for ii in ['Nscore','Escore','Oscore','Ascore','Cscore','Impulsive','SensationSeeking']:
    
    for i in ["stimulant","depressant","hallucinagen"]:
    
        sns.lineplot(df[ii],df[i])
        plt.legend(["stimulant","depressant","hallucinagen"])
       
        plt.ylabel('drug use likelihood')    
    plt.show()
    
```

```python
for i in ['Nscore','Escore','Oscore','Ascore','Cscore','Impulsive','SensationSeeking']:
    sns.scatterplot(df.groupby(i)[["stimulant","depressant","hallucinagen"]].mean().unstack([str(i)]))
```
